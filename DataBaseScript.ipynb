# [dbo].[usp_VentaCrear]

```sql
/****** Object:  StoredProcedure [dbo].[usp_VentaCrear]    Script Date: 8/13/2025 11:51:20 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE   PROCEDURE [dbo].[usp_VentaCrear]
    @Cliente NVARCHAR(120),
    @Items   dbo.VentaItemInput READONLY,
    @VentaId INT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    SET XACT_ABORT ON;

    BEGIN TRY
        BEGIN TRAN;

        -- 1) Validar stock
        IF EXISTS (
            SELECT 1
            FROM @Items i
            JOIN dbo.Productos p ON p.Id = i.ProductoId
            WHERE i.Cantidad > p.Stock
        )
        BEGIN
            THROW 50001, 'Stock insuficiente para uno o más productos.', 1;
        END

        -- 2) Crear venta (Total 0)
        INSERT INTO dbo.Ventas (Cliente, Total)
        VALUES (@Cliente, 0);

        SET @VentaId = SCOPE_IDENTITY();

        -- 3) Insertar ítems (si PrecioUnitario es NULL, usar p.Precio)
        INSERT INTO dbo.VentaItems (VentaId, ProductoId, Cantidad, PrecioUnitario)
        SELECT
            @VentaId,
            i.ProductoId,
            i.Cantidad,
            ISNULL(i.PrecioUnitario, p.Precio)
        FROM @Items i
        JOIN dbo.Productos p ON p.Id = i.ProductoId;

        -- 4) Descontar stock
        ;WITH mov AS (
            SELECT ProductoId, SUM(Cantidad) Cantidad
            FROM @Items
            GROUP BY ProductoId
        )
        UPDATE p
        SET p.Stock = p.Stock - m.Cantidad
        FROM dbo.Productos p
        JOIN mov m ON m.ProductoId = p.Id;

        -- 5) Recalcular total
        UPDATE v
        SET v.Total = t.Total
        FROM dbo.Ventas v
        CROSS APPLY (
            SELECT SUM(vi.Cantidad * vi.PrecioUnitario) AS Total
            FROM dbo.VentaItems vi
            WHERE vi.VentaId = @VentaId
        ) t
        WHERE v.Id = @VentaId;

        COMMIT;
    END TRY
    BEGIN CATCH
        IF XACT_STATE() <> 0 ROLLBACK;
        THROW;
    END CATCH
END
GO
```
